# Inoperability Input-Output Model

This notebook provides an example for how to carry out IIM analysis.

```python
import os
import iim.iim as iim
import iim.io as iim_io
import matplotlib.pyplot as plt
```

### Input data

```python
fname = os.path.join("..", "examples", "ssb_io.csv")
psector = ["RD"]
cvalue = [0.1]
mode = "IO"

model = iim.IIM(fname, psector, cvalue, mode)
sectors = model.get_sectors()
```

### Dependency index

```python
delta = model.dependency()
iim_io.plot(sectors, delta, "Industry sector", "Dependency index")
```

### Influence gain

```python
rho = model.influence()
iim_io.plot(sectors, rho, "Industry sector", "Influence gain")
```

### Inoperability

```python
q = model.inoperability()
iim_io.plot(sectors, q, "Industry sector", "Inoperability")
```
